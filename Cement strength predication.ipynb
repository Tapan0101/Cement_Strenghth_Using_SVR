### Data trasformation

```python
X = df.drop('Concrete compressive strength(MPa, megapascals) ',axis=1)
y = df['Concrete compressive strength(MPa, megapascals) ']
```

```python
for col in X:
    X[col] =  np.log(X[col]+2)
```

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
pn = 1
for col in X:
    plt.subplot(3,3,pn)
    sns.distplot(X[col])
    plt.xlabel(col,fontsize=10)
    pn = pn+1
plt.show()
```

```python
## lets cheack outlier
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
pn = 1
for col in X:
    plt.subplot(3,3,pn)
    sns.boxplot(X[col])
    plt.xlabel(col,fontsize=10)
    pn = pn+1
plt.show()
```

```python
plt.figure(figsize=(15,10))
pn =1 
for col in X:
    plt.subplot(3,3,pn)
    sns.scatterplot(X[col],y)
    plt.xlabel(col,fontsize = 10)
    pn = pn+1
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X.corr(),annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.svm import SVR
```

```python
reg = SVR()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
para = {'C':[0.5,0.2,1,2 ,3, 4],'gamma':[0.5,0.2,1,2 ,3, 4]}
```

```python
grid = GridSearchCV(reg,param_grid=para,cv=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
reg_tune = SVR(C=4,gamma=0.2)
reg_tune.fit(X_train,y_train)
reg_tune.score(X_train,y_train)
```

```python
reg_tune.score(X_test,y_test)
```
